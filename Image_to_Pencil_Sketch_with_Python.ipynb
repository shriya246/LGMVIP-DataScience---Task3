```python
import cv2
import matplotlib.pyplot as plt
import seaborn as sns
```

**Reading the image**

```python
img = cv2.imread("Anime.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.figure(figsize = (6,6))
plt.imshow(img)
plt.axis("off")
plt.title("\n Anime Original Image\n")
plt.show()
```

**Converting The Original Image To GreyScale Image**

```python
grey_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(6,6))
plt.imshow(grey_img,cmap="gray")
plt.axis("off")
plt.title("\nAnime Gray Scale Image\n")
plt.show()
```

**Inverting the greyscale image**

```python
inverted_img = cv2.bitwise_not(grey_img)
plt.figure(figsize=(6,6))
plt.imshow(inverted_img,cmap ="gray")
plt.axis("off")
plt.title("\n Anime Inverted Image\n")
plt.show()
```

```python
#Blurring the Image using Gaussian Function in OpenCv
smoothened_img = cv2.GaussianBlur(inverted_img, (21,21), sigmaX=0, sigmaY=0)
smoothened_img
```

```python
plt.figure(figsize=(6,6))
plt.imshow(smoothened_img , cmap = "gray")
plt.axis("off")
plt.title("\nAnime Smoothened Image\n")
plt.show()
```

**Inverting The Blurred Image**

```python
final_output = cv2.divide(grey_img, 255 - smoothened_img, scale=255)
plt.figure(figsize=(6,6))
plt.imshow(final_output,cmap="gray")
plt.axis("off")
plt.title("\nAnime Pencil Sketch Image\n")
plt.show()
```

```python
#Comparison Between Original & Sketch Image
plt.figure(figsize=(20,20))
plt.subplot(1,5,1)
plt.imshow(img)
plt.axis("off")
plt.title("\n Anime Naruto - Original Image\n")
plt.subplot(1,5,2)
plt.imshow(final_output,cmap="gray")
plt.axis("off")
plt.title("\n Anime Naruto- Pencil Sketch\n")
plt.show()
```
